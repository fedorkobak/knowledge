# Jinja Templates

Jinja templates offers a flexible way to parameterise text documents. They allows various documents to be created from a template using given data.

Originally, jinja is a python package that implements templating logic, but it used in various systems as a templating language. Thus, this page considers the template syntax, not the python api. 

Check more on templating in the [Template Designer Documentation](https://jinja.palletsprojects.com/en/stable/templates/).

```python
from jinja2 import Template
```

## Control statements

The control statement determines the flow of the pattern.

Control statemens in jinja are wrapped in the pattern `{% <statement> %}`.

This is a list of control statements with short description:

| Statement          | Description                             |
| ------------------ | --------------------------------------- |
| `if / elif / else` | Conditional branching                   |
| `for`              | Iterate over a sequence                 |
| `for ... else`     | Loop with fallback if sequence is empty |
| `break`            | Exit loop early                         |
| `continue`         | Skip current iteration                  |
| `set`              | Assign variable                         |
| `macro`            | Define reusable template function       |
| `include`          | Insert another template                 |
| `extends`          | Inherit from a base template            |
| `block`            | Define overridable section in template  |
| `with`             | Create scoped variables                 |
| `raw`              | Prevent template rendering inside block |
| `filter`           | Apply filter to a block of content      |

Check detailed description in the [List of Control Structures](https://jinja.palletsprojects.com/en/stable/templates/#list-of-control-structures).

---

The following cell shows the use of the `for` control statement.

It wraps the elements of the given list with `!` symbol.

```python
template = Template("{% for i in values %}!{{ i }}!{% endfor %}")
template.render(values=[10, "hello", 3.14])
```

## Expressions

In jinja, expressions print the template output. Expressions are wrapped in the pattern `{{ <expression> }}`. In an expression you can use:

- [Literals](https://jinja.palletsprojects.com/en/stable/templates/#literals): representations for Python objects as strings and numbers: `{{ "Hello world" }}`, `{{ 200 }}`, `{{ [20, "string", 3.14] }}`.
- [Math](https://jinja.palletsprojects.com/en/stable/templates/#math): you can perform some calculation inside the expression: `{{ 1 + 1 }}`, `{{ 3 ** 2 }}`, `{{ '=' * 80 }}`.
- [Comparisons](https://jinja.palletsprojects.com/en/stable/templates/#comparisons).
- [Logic](https://jinja.palletsprojects.com/en/stable/templates/#logic).
- [Other](https://jinja.palletsprojects.com/en/stable/templates/#other-operators).
- [If Expression](https://jinja.palletsprojects.com/en/stable/templates/#if-expression): python one-line if statements could be used inside the jinja expresssion: `{{ "some literal" if var == 10 }}`.
- [Python methods](https://jinja.palletsprojects.com/en/stable/templates/#python-methods): you can call methods from python objects if they have corresponding type `{{ some_string_var.capitalize() }}`.

Check more the [Expressions](https://jinja.palletsprojects.com/en/stable/templates/#expressions) section of the official documentation.

---

The following cell defines the expression that uses some possibilities.

```python
template = Template(
    "{{ 'correct'.capitalize() if val1 == 10 and val2 == 'hello' else 'incorrect' }}"
)
```

This is the render for values that pass the `if` statement.

```python
template.render({"val1": 10, "val2": 'hello'})
```

It's almost the same, but it follows a different branch of logic.

```python
template.render({"val1": 11, "val2": 'hello'})
```
