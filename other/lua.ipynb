## Functions

There are two syntaxes for defining a function, both of which are represented in the following example:

```lua
function hello(name)
    print("Hello!", name)
end

hello("Lua")
```

```lua
hello2 = function(name)
    print("Hello!", name)
end

hello2("Lua2")
```

## Tables

Lua offers the `table` datatype to store data. This object can represent both sequences and mappings.

---

The following cell defines the `table1` as table:

```lua
table1 = {
    'the first element',
    key1 = 'value1',
    ['key2'] = 'value2'
}
table1
```

Here are table that contains:

- The "the first element" string as the first element of the sequence. 
- Under the `key1` hides value `value1`.
- Under the `key2` hides value `value2`.

By accessing the first index of the table, you retrieve value that is defined as an element of the table:

```lua
table1[1]
```

There are different ways to access the values under the mapping:

```lua
table1["key1"]
```

```lua
table1.key2
```

### Length

To get the length of the lua table use `#` operator.

**Note** it only counts sequential elements of the table; it ignores mappings.

## Loops

There are two classical loops in lua: `for` and `while`.

---

The following cell provides an option to define a for loop that iterates over the index:

```lua
for index=1, 5 do
    print(index)
end
```

There is another option for the  `for` loop. The syntax `for <variables> in <iterator triplet>` could be used. The returns the objects at each iteration, which would be assiged to the `variables`.

The following cell demonstrates this kind of syntax using the `ipairs` function, which returns the iterator triplet that iterates over the sequence stored in the table:

```lua
for index, val in ipairs({20, 30, 2, 1, key=value}) do
    print(index, val)
end
```

The syntax of the while loop is really straight:

```lua
i = 1
while i < 5 do
    print(i)
    i = i + 1
end
```
