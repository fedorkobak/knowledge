## Tables

Lua offers the `table` datatype to store data. This object can represent both sequences and mappings.

---

The following cell defines the `table1` as table:

````lua
table1 = {
    'the first element',
    key1 = 'value1',
    ['key2'] = 'value2'
}
table1
````

Here are table that contains:

- The "the first element" string as the first element of the sequence. 
- Under the `key1` hides value `value1`.
- Under the `key2` hides value `value2`.

By accessing the first index of the table, you retrieve value that is defined as an element of the table:

````lua
table1[1]
````

There are different ways to access the values under the mapping:

````lua
table1["key1"]
````

````lua
table1.key2
````

### Methods

There is no such thing as a 'method' in lua. However, you can store functions in the tables and pass the table itself to them as an argument.

Lua has a special operator for this: column `:`:

**Define the method** using syntax:

```lua
function table_name:method_name(param1, param2, param3)
...
end
```

The key feature of the method is that it automatically  defines the `self` parameter, which you can use in the function body.

In Lua's logic, it is formally still a function with an implicitly defined `self` parameter. Therefore, if you call in the usual way using the `.` operator, you must implicitly an argument for `self`. To simplify that `:` can be used to assign to the method of the table:

```lua
table_name:method_name(arg1, arg2, arg3)
```

---

The following cell shows the table that has:

- `column_method` defined through `:` operator.
- `regular_functions` defined through `.` operator.

Both of them are trying to print the `self` object.

````lua
some_table = { value = 'the value' }

function some_table:column_method()
    print(self)
end

function some_table.regular_function()
    print(self)
end
````

First, assign to the `column_method` through `:`:

````lua
some_table:column_method()
````

This is the regular behaviour of the method. The `self` simply refers to the `some_table` itself.

However, the same result could be achieved explicitly:

````lua
some_table.column_method(some_table)
````

And an attempt to call the `regular_function` through the `:` operator:

````lua
some_table:regular_function()
````

Even though the function was called with the `:` operator, the `self` argument cannot be accessed in the function body as it was not defined.

## Flow control

The flow control constructions available in lua are listed in the following table:

| Construct                                          | Description                                     |
| -------------------------------------------------- | ----------------------------------------------- |
| `if ... then ... end`                              | Conditional execution.                          |
| `if ... then ... else ... end`                     | Conditional execution with alternative branch.  |
| `if ... then ... elseif ... then ... else ... end` | Multi-branch conditional.                       |
| `while ... do ... end`                             | Pre-condition loop.                             |
| `repeat ... until`                                 | Post-condition loop (runs at least once).       |
| `for i = start, stop [, step] do ... end`          | Numeric loop.                                   |
| `for ... in ... do ... end`                        | Generic loop using iterators.                   |
| `break`                                            | Exits the innermost loop.                       |
| `goto label`                                       | Jumps to a label within the same function.      |
| `::label::`                                        | Defines a label for `goto`.                     |
| `return`                                           | Exits a function and optionally returns values. |
| `do ... end`                                       | Creates a local scope block.                    |


Check mode detailed explanation with some examples in the dedicated [Flow control](lua/flow_control.ipynb) page.

## Standard library

Check the [Standard libraries](https://www.lua.org/manual/5.5/manual.html#6) reference page.

The following talbe show the most essential functions of the lua standard library:

| Name       | Description                                                                             |
| ---------- | --------------------------------------------------------------------------------------- |
| `print`    | Prints values to standard output, converting them to strings.                           |
| `type`     | Returns the type of a value (`"nil"`, `"number"`, `"string"`, `"table"`, etc.).         |
| `pairs`    | Returns an iterator for traversing all key-value pairs in a table.                      |
| `ipairs`   | Returns an iterator for traversing the integer-indexed elements of an array-like table. |
| `next`     | Returns the next key-value pair in a table; used internally by `pairs`.                 |
| `tonumber` | Converts a value to a number if possible.                                               |
| `tostring` | Converts a value to its string representation.                                          |
| `pcall`    | Calls a function in protected mode, catching errors instead of raising them.            |
| `require`  | Loads and returns a module, caching it for future use.                                  |
| `error`    | Raises an error and stops normal execution unless caught by `pcall` or `xpcall`.        |

## Modules

The modles in lua can be imported by standard library `require` function.

The module is simply a lua file that returns a value. Typically, this is table named `M` that contains references to all the objects the module provides.

---

Consider the module `lua_files/example.lua`:

```lua
M = {}

M.some_function = function()
    print("This is function from module")
end

return M
```

The code that imports that module:

````lua
example_module = require("lua_files.example")
example_module
````

The usage of `some_function` has the expected behaviour:

````lua
example_module.some_function()
````
