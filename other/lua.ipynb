## Tables

Lua offers the `table` datatype to store data. This object can represent both sequences and mappings.

Check more details in the specific [tables](lua/tables.ipynb) page.

---

The following cell defines the `table1` as table:

````lua
table1 = {
    'the first element',
    key1 = 'value1',
    ['key2'] = 'value2'
}
table1
````

Here are table that contains:

- The "the first element" string as the first element of the sequence. 
- Under the `key1` hides value `value1`.
- Under the `key2` hides value `value2`.

By accessing the first index of the table, you retrieve value that is defined as an element of the table:

````lua
table1[1]
````

There are different ways to access the values under the mapping:

````lua
table1["key1"]
````

````lua
table1.key2
````

## Flow control

The flow control constructions available in lua are listed in the following table:

| Construct                                          | Description                                     |
| -------------------------------------------------- | ----------------------------------------------- |
| `if ... then ... end`                              | Conditional execution.                          |
| `if ... then ... else ... end`                     | Conditional execution with alternative branch.  |
| `if ... then ... elseif ... then ... else ... end` | Multi-branch conditional.                       |
| `while ... do ... end`                             | Pre-condition loop.                             |
| `repeat ... until`                                 | Post-condition loop (runs at least once).       |
| `for i = start, stop [, step] do ... end`          | Numeric loop.                                   |
| `for ... in ... do ... end`                        | Generic loop using iterators.                   |
| `break`                                            | Exits the innermost loop.                       |
| `goto label`                                       | Jumps to a label within the same function.      |
| `::label::`                                        | Defines a label for `goto`.                     |
| `return`                                           | Exits a function and optionally returns values. |
| `do ... end`                                       | Creates a local scope block.                    |


Check mode detailed explanation with some examples in the dedicated [Flow control](lua/flow_control.ipynb) page.

## Modules

The important builtin facilities to manupulate with modules:

- The `package` table contains the attributes that determine module-related behaviour.
- The `require` function loads the module.

The module is simply a lua file that returns a value. Typically, this is table named `M` that contains references to all the objects the module provides.

Check more in the [Modules](https://www.lua.org/manual/5.5/manual.html#6.4) section of the lua manual.

---

Consider the module `lua_files/example.lua`:

```lua
M = {}

M.some_function = function()
    print("This is function from module")
end

return M
```

The code that imports that module:

````lua
example_module = require("lua_files.example")
example_module
````

The usage of `some_function` has the expected behaviour:

````lua
example_module.some_function()
````

Now consider the `package` table:

````lua
print(type(package))
````

As key-values it contains all the modules that have been loaded into the environment:

````lua
for k, _ in pairs(package.loaded) do
    print(k)
end
````

For example, the file `lua_files.example` that was imported manually. The following cell accesses the module through `package.loaded` table:

````lua
package.loaded["lua_files.example"]
````

The `searchers` attribute of the `package` contains a set of functions that an be used to search for modules. Consequently, you can implement your own module-looking strategy simply by adding a new functions to this table:

````lua
package.searchers
````

### Searchpath

The `packages.searchpath(name, path [, sep [, rep]])` function searches for a specified files.

The `name` determines the names of the file.

The `path` determines set of patterns where `name` would be substituted separted by the `;`.

The first case in which the value of the `name` is substitued into the value of the variable path, resulting in a path to an existing file that can be ingested, returns the path to this file.

---

The following cell roughly shows how lua searches for the `.lua` files.

````lua
package.searchpath('example', './lua_files/?.lua')
````

If the name or pattern are incorrect, the `nil` value and a message describing the issue are returned.

````lua
print(package.searchpath('some_other', './lua_files/?.lua'))
print(package.searchpath('example', './lua_files/?'))
````
