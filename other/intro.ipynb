## Agile

The Agile is to approach for software development that relies on the following principles: you have to give more value to

- Individuals and interactions over processes and tools.
- Working software over comprehensive documentation
- Communicating with customers over negotiating contracts.
- Responding to change over following a plan.

There two main frameworks that follows Agile principle: Scrum and Kanban.

### Scrum

In Scrum, work involves iterative improvements to the software. Each iteration takes from one to four weeks and is called a *sprint*. There are:

- **Events**: periodical meetings or other organisational ceremonies that are usually tied to a particular phase of the software development.
- **Artifacts**: the things to be done. Separated into few groups typically: *backlog*, *to do*, *in process* and *done*.
- **Roles**: 
    - *Product owner*: person responsible for communication with the customer.
    - *Scrum master*: person that checks that scrum rules are followed.
    - *Scrum team*: people who complete the tasks.

### Kanban

Kanban just takes from scrum the groups of tasks: *to do*, *doing* and *done*.


## Markdown

Markdown is a special text formatting language. The overview:

* **Headings:** `# H1`, `## H2`, `### H3`
* **Emphasis:** `*italic*`, `**bold**`, `~~strikethrough~~`
* **Lists:**

  * Unordered: `- item` or `* item`
  * Ordered: `1. item`
* **Links:** `[text](https://example.com)`
* **Images:** `![alt text](image.png)`
* **Code:**

  * Inline: `` `code` ``
  * Block:

    ````markdown
    ```python
    print("hello")
    ```
    ````
* **Blockquotes:** `> quote`
* **Tables:**

  ```markdown
  | A | B |
  |:-|:-:|
  | 1 | 2 |
  ```
* **Horizontal line:** `---` or `***`
* **Inline HTML:** `<b>bold</b>`

It’s designed to be readable as plain text and easily converted to HTML.

### Links

- Any text that begins with `https://` will be interpreted as a link and will be clickable.
- To create a hyperlink use the syntax `[represented text](<addres>)`.
- It's not alway convenient to put the address right after text. You can define the hyperlink as `[my link][1]` and specify the address later in the text as `[1]: <address>`. This approach is called **refence-style link**.

## Networks

This section considers the principles of networking.

There are a few important protocols that describe the rules of communication within the network.

There is a set of rules for transfering data in a network, typically called protocols. The following typical protocols:

- HTTP (HyberText Transfer Protocol): The protocol used by applications to transfer data. Applications are responsible for preparing messages in the corresponding format and for interpreting the messages they receive.
- TCP (Transmission Control Protocol): Transport layer.
- IP (Internet Protocol): Network layer.
- Network-specific link interface: Data link layer.
- Physical network hardware: Physical layer.

## JSON schema

JSON schema is a declarative protocol to specify the format of the JSON documents. 

For more check:

- The [Official website](https://json-schema.org/).
- For implemetations of JSON schema visit [tools](https://json-schema.org/tools?query=&sortBy=name&sortOrder=ascending&groupBy=toolingTypes&licenses=&languages=&drafts=&toolingTypes=&environments=&showObsolete=false&supportsBowtie=false):
    - The python implementation is [jsonschema](https://github.com/python-jsonschema/jsonschema) package.
    - The python package with CLI [check-jsonschema](https://pypi.org/project/check-jsonschema/).
- The [JSON Schema](json_schema.ipynb) page.

---

The following cell specifies a schema in JSON format with the attributes "price" and "name". The "price" attribute must be an integer value, and "name" is declared to be a string.

`````bash
cat << EOF > /tmp/schema.json 
{
    "type" : "object",
    "properties" : {
        "price" : {"type" : "number"},
        "name" : {"type" : "string"}
    }
}
EOF
`````

The `check-jsonschema` verifies that the files passed correspond to the schema specified by the `--schemafile` argument.

`````bash
cat << EOF > /tmp/object.json
{"name": "Eggs", "price": 43.99}
EOF
check-jsonschema --schemafile /tmp/schema.json /tmp/object.json
`````

Validation is successful. However, the attempt to pass a number as a string to the `price` in the following cell fails.

`````bash
cat << EOF > /tmp/object.json
{"name" : "Eggs", "price" : "34.99"}
EOF
check-jsonschema --schemafile /tmp/schema.json /tmp/object.json
`````

## Unicode

UTF-8 is the most popular method for representing text on computer. UTF-8 encoding is usually written as `U+uvwxyz`, where `uvwxyz` is 6-digit hex number.

Depending on how far the symbol if from the beginning of the table, it can take from one to four bytes.

Check the [Common References for Unicode Standard](https://unicode.org/reports/tr41/tr41-36.html). It contains various technical details about the Unicode specification. This sidte considers the features of the latest unicode standard if other is not mentioned specifically.

Sometimes it can be usefull to find the specific character and data related to it. Check the [Unicode Character Database](https://www.unicode.org/ucd/).

---

### Layers of character


There are three layers that defines each character in unicode:

- **Abstract character**: the single piece of information. The concept itself.
- **Code point**: the numerical value assigned to that concept.
- **Glyph**: the spefic "image" that represents the code point. Each font have it's own glyphes.

---

For example, consider the case where to the same abstract character "latin small letter e with acute" has two representations:

`````bash
printf "\u00E9"
printf "\u0065\u0301"
`````

First uses just code point `U+00E9`, second is composed from `U+0065` and `U+0301`, but the abstract character is the same:лллл

The following example illustrates the case in which two code point have the same abstract character due to compatibilities issues:

`````bash
printf "\u212B"
printf "\u00C5"
`````

### Normalization forms

The same human-readable text can be represented in Unicode in variety of ways. There are normal forms of Unicode text that define a single way for text to be represented.

Check more in the [Unicode normalization forms](https://unicode.org/reports/tr15/).

---

The following cell shows how the same symbols that can be encoded differently in Unicode.

`````bash
printf "\u00C7"
printf "\u0043\U0327"
`````

## Data build tool

Data Build Tool (dbt) automates and standartises the project associated with transforming data.

In DBT, a project is a set of configuration files that define how data should be processed.

A dbt project cab be run in different ways, but the most of these are wrappers around the CLI. There are two CLIs:

- `dbt-core` python package.
-  dbt fustion (`dbtf`) the modern interface that is distributed as system package. Check the [installation guide](https://docs.getdbt.com/docs/local/install-dbt?version=2.0#installation).

The dbt works with a specific data storage system like: snowflake, postgres, clickhouse ect. You can define the credentials required to access the system in the "profile". The profiles are defined in the file `/.dbt/profiles.yml`.

For more check the dedicated [DBT](dbt.ipynb) page.

---

To create the project, simply run the `dbtf init`. The interface will then prompt you for all the necessary configuration inputs. 

As this site's structure follows the style of a Jupyter notebook, it is not possible to display the command-line interface (CLI) prompts. However, you can create the profile manually and build a project based on it.

For example profile may look like this:

```yaml
knowledge:
  target: dev
  outputs:
    dev:
      type: postgres
      port: 5432
      database: postgres
      schema: public
      threads: 16
      host: localhost
      user: postgres
      password: postgres
```

The process of creating the project:

`````bash
rm -rf /tmp/new_data && mkdir /tmp/new_data
cd /tmp/new_data

dbtf init --project-name knowledge --profile knowledge
`````

**Note**: you must start the postgreSQL database using correct credentials.
