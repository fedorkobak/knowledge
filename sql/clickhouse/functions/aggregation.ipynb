# Aggregation

The idea of the aggregation function in ClickHouse works in same way as in other SQL dialects, combining sets of values into a single value.

## Combinators

Combinators are special suffixes for aggregation functions that modify their behaviour. You can add an arbitrary number of combinators an aggregation function to achieve a desiered outcome.

Check the [Aggregate function combinators](https://clickhouse.com/docs/sql-reference/aggregate-functions/combinators) page to achieve a specific behavior.

---

The following cell generates an example using arrays and groups, applying the `sum` function with different combinators.

```bash
--ClickHouse
WITH combin_example AS (
    SELECT *
    FROM VALUES(
        'my_array Array(UInt8), group TEXT',
        ([1, 2, 3, 4], 'a'),
        ([3, 2, 1], 'a'),
        ([5, 4, 3], 'b')
    )
)
SELECT
    sumArray(my_array) AS sumArray,
    sumArrayIf(my_array, "group" = 'a')  AS sumArray_for_a
FROM combin_example;
```

The first column of the output is `sumArray`, which instructs ClickHouse to treat arrays as sequences of values. The second column, in addition to that, selects only that arrays that correspon to "a" group.
