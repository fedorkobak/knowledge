# Strings

This page looks at the features of the string processing functions.

## Split By

There is a set of functions that allows you to split the strings and put the results into an array:

- [`splitByChar`](https://clickhouse.com/docs/sql-reference/functions/splitting-merging-functions#splitByChar).
- [`splitByString`](https://clickhouse.com/docs/sql-reference/functions/splitting-merging-functions#splitByString).
- [`splitByRegexp`](https://clickhouse.com/docs/sql-reference/functions/splitting-merging-functions#splitByChar).

---

The options for splitting the string are shown in the following cell:

```sql
--ClickHouse
SELECT
    col,
    splitByChar('|', col),
    splitByString('<|>', col),
    splitByRegexp('<.*>', col)
FROM values(
    'col String',
    ('helo<|>this<|>is'),
    ('extract<some>this<complex>text')
)
```

**Note** if the splitting pattern is at the sides of the string empty strings will be included in the output:

```sql
--ClickHouse
SELECT
    splitByChar('|', '|the|');
```

## like/ilike

Possibly the most popular functions for searching for a string that corresponds to a particular pattern are `like` and `ilke`. They are the same, but `ilike` is case-insensitive.

The following ideas are important about this functions:

- The `%` defines the arbitrary amount of symbols.
- The `_` defines the single symbol.
- You can specify the escape character, by default use `\`.
- There are corresponding operators `<col> like <pattern>` and `<col> ilike <pattern>`, respictively.

Check more in the corrspoding section of documentation [like](https://clickhouse.com/docs/sql-reference/functions/string-search-functions#like) and [ilike](https://clickhouse.com/docs/sql-reference/functions/string-search-functions#ilike).

---

The following cell shows the result of searching for the same pattern using `like` and `ilike`:

```sql
--ClickHouse
SELECT
    col, 
    like(col, 'foo%'),
    ilike(col, 'foo%')
FROM values(
    'col String',
    ('foo'),
    ('bar'),
    ('FooBar')
);
```
