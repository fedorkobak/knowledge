# Fucntions

This page considers fucntions awailable in clickhouse.

## Regular

The list of regular fucntions specific for ClickHouse.


| Function                         | Description                                                                 | Typical Use Case                                  |
|----------------------------------|-----------------------------------------------------------------------------|---------------------------------------------------|
| arrayMap(f, arr)                | Applies function `f` to each element of array                               | Vectorized transformations                        |
| arrayFilter(f, arr)             | Filters array elements by condition                                         | Array filtering                                   |
| arrayReduce(agg, arr)           | Applies aggregation function to array                                       | Aggregate inside arrays                           |
| arrayJoin(arr)                  | Expands array into multiple rows                                            | Explode arrays (like UNNEST)                      |
| arrayZip(arr1, arr2, ...)       | Combines arrays element-wise into tuples                                    | Align multiple arrays                             |
| arrayEnumerate(arr)             | Returns indices of array elements                                           | Position tracking                                 |
| map(key1, val1, ...)            | Creates a map (key-value structure)                                         | Structured data                                   |
| mapKeys(map)                    | Returns keys of map                                                         | Inspect map                                       |
| mapValues(map)                  | Returns values of map                                                       | Inspect map                                       |
| mapGet(map, key)                | Returns value for key                                                       | Lookup in map                                     |
| tuple(x, y, ...)                | Creates tuple                                                               | Multi-value expressions                           |
| tupleElement(tuple, i)          | Extracts element by index                                                   | Access tuple values                               |
| JSONExtract(json, type)         | Extracts JSON field with type                                               | JSON parsing                                      |
| JSONExtractString(json, key)    | Extracts string value from JSON                                             | JSON access                                       |
| JSONExtractInt(json, key)       | Extracts integer from JSON                                                  | JSON access                                       |
| toStartOfDay(dt)                | Rounds datetime to start of day                                             | Time bucketing                                    |
| toStartOfHour(dt)               | Rounds datetime to start of hour                                            | Time aggregation                                  |
| toStartOfInterval(dt, interval) | Rounds datetime to custom interval                                          | Flexible time windows                             |
| dateDiff(unit, t1, t2)          | Difference between dates                                                    | Time calculations                                 |
| addDays(dt, n)                  | Adds days to date                                                           | Date arithmetic                                   |
| now()                           | Current timestamp                                                           | Real-time queries                                 |
| rand()                          | Random number                                                               | Sampling / testing                                |
| cityHash64(x)                   | Fast non-cryptographic hash                                                 | Hashing for joins / sharding                      |
| sipHash64(x)                    | Deterministic hash                                                          | Consistent hashing                                |
| reinterpretAsUInt32(x)          | Reinterprets bytes as another type                                          | Low-level transformations                         |
| assumeNotNull(x)                | Treats nullable as non-nullable                                             | Performance optimization                          |
| if(cond, a, b)                  | Conditional expression                                                      | Faster alternative to CASE                        |
| multiIf(cond1, val1, ..., else) | Multiple conditions                                                         | Compact CASE replacement                          |
| greatest(a, b, ...)             | Maximum of values                                                           | Comparisons                                       |
| least(a, b, ...)                | Minimum of values                                                           | Comparisons                                       |
| runningDifference(x)            | Difference between current and previous row                                 | Time series analysis                              |
| neighbor(x, offset)             | Access nearby row value                                                     | Window-like behavior without window functions

## Aggregation

The idea of aggregation functions in ClickHouse generally the same as in other DBMSs. Here, we mainly consider the functions that are specific to ClickHouse.


Following cell show the most usefull clickhouse aggregation functions.

| Function                | Description                                                                 | Typical Use Case                                  |
|------------------------|-----------------------------------------------------------------------------|---------------------------------------------------|
| any()                  | Returns an arbitrary value from the group                                   | Fast approximation when exact value not needed    |
| anyLast()              | Returns the last encountered value                                          | When order matters (with ORDER BY)                |
| argMax(value, weight)  | Returns `value` corresponding to max `weight`                               | Get row value for max metric                      |
| argMin(value, weight)  | Returns `value` corresponding to min `weight`                               | Get row value for min metric                      |
| groupArray()           | Aggregates values into an array                                             | Collect all values per group                      |
| groupUniqArray()       | Array of unique values                                                      | Deduplicated list aggregation                     |
| groupArrayInsertAt()   | Inserts values into array at specific positions                             | Sparse array construction                         |
| groupBitmap()          | Builds bitmap from values                                                   | Efficient set representation                      |
| bitmapOr()             | Aggregates bitmaps with OR                                                  | Merge sets efficiently                            |
| bitmapAnd()            | Aggregates bitmaps with AND                                                 | Intersection of sets                              |
| uniq()                 | Approximate distinct count (fast)                                           | High-performance cardinality estimation           |
| uniqExact()            | Exact distinct count                                                        | When precision is required                        |
| uniqCombined()         | Hybrid approximate distinct                                                 | Balance between speed and accuracy                |
| uniqHLL12()            | HyperLogLog-based distinct count                                            | Large-scale cardinality estimation                |
| quantile()             | Approximate quantile                                                        | Median / percentiles                              |
| quantileExact()        | Exact quantile                                                              | Precise percentile calculation                    |
| quantileTDigest()      | TDigest-based quantile                                                      | Accurate distribution estimation                  |
| topK(N)                | Returns top N frequent values                                               | Heavy hitters / frequency analysis                |
| sumMap()               | Aggregates maps by summing values per key                                   | Aggregation over key-value pairs                  |
| avgWeighted()          | Weighted average                                                            | Metrics with weights                              |
| corr()                 | Computes correlation coefficient                                            | Statistical analysis                              |
| covarPop()             | Population covariance                                                      | Statistical analysis                              |
| skewPop()              | Skewness                                                                   | Distribution shape analysis                       |
| kurtPop()              | Kurtosis                                                                   | Distribution tail analysis                        |


The important concept here is **combinators**: suffixes that can be added to the aggregation functions to modify their behaviour in a specific way.
