# Configuration

## Files

```bash
#init
#file new_config.xml
<clickhouse>
    <logger>
        <level>test</level>
    </logger>
</clickhouse>
```

```bash
docker run -d --rm --name configuration_click \
    -v ./new_config.xml:/etc/clickhouse-server/config.d/new_config.xml \
    clickhouse:latest
```

```bash
docker exec -it configuration_click clickhouse-client \
    -q "select name, value from system.server_settings where name='logger.level'"
```

```bash
docker exec -it configuration_click ls /etc/clickhouse-server
```

```bash
docker stop configuration_click
```
