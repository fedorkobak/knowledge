# Configuration

This page considers the process and aspects of the clickhouse configuration.

It uses special docker container where this notebook have to be started: `clickhouse`.

## Files

Clickhouse stores the configuration files in the directory `/etc/clickhouse-server`.

There is a default configuration file `config.xml`. While the configuration can be changed directly in the `config.xml`, it is recomended to put new configuration in the separate files in the `/etc/clickhouse-server/config.d` folder. The configurations would be merged according [specific rools](https://clickhouse.com/docs/operations/configuration-files#merging).

Check more about configuration files of the clickhouse in the [Configuration files](https://clickhouse.com/docs/operations/configuration-files) section of the documentation.

---

The following cells define the simple configuration file and pushes it to the clickhouse.

```bash
#init
#file new_config.xml
<clickhouse>
    <logger>
        <level>test</level>
    </logger>
</clickhouse>
```

```bash
cp ./new_config.xml /etc/clickhouse-server/config.d/new_config.xml
service clickhouse-server restart &> /dev/null
```

The restarted clickhouse instance the option specified in config:

```bash
clickhouse-client \
    -q "select name, value from system.server_settings where name='logger.level'"
```

And general structure of the confugration files in clickhouse:
