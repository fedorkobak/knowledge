# PostgreSQL

PostgreSQL is a popular relational database. It is a relatively universal database system that can be used in numerous cases. This page considers the aspects of working with PostgreSQL.

## Databases

Single postgreSQL instance can contain several databases. Check the available databases in the `pg_database` table.

---

The contents of the `pg_database` table are shown in the following cell.

```sql
--postgreSQL
SELECT * FROM pg_database;
```
