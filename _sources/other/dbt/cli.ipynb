# CLI

This page looks at the commands available via `dbt-core` and `dbtf`. They are the primary ways to manipulate the dbt project.

## Initialization

You can initialise the project (create all necessary for dbt strcutures) by using the `dbt(f) init` command. There are slightly different interfaces for `dbt-core` and fusion.

In general, you can simply `dbt init` or `dbtf init` and CLI will ask you promt you for all the necessary parameters. However, due to restrictions in the Jupyter notebook, the project name and profile must be specified explicetely.

---

The following cell initialises the `dbt-core` project.

```bash
#init
dbt init init_example --profile knowledge -q
tree
```

Next code show the same operation but for fustion. **Note** unlike the `dbt-core`, fusion requires name of the project to be specified with named argument `--project-name`.

```bash
#init
dbtf init --project-name knowledge --profile knowledge -q
tree
```

## Compilation

Compilation is a proccess of converting DBT language into scripts specific to the target.

The `dbtf compile` initiates the compilation. It produces the outputs to the `target` folderлл.

---

The following cell inititalises the example project.

```bash
#init
dbtf init --project-name knowledge --profile knowledge -q
cd knowledge
```

The template with jinja templates and compilation.

```bash
# file knowledge/models/script.sql
CREATE TABLE 
    {% if target.name == 'dev' %} 
        dev_table
    {% else %} 
        prod_table
    {% endif %}
(
    some_value INTEGER
)
```

```bash
dbtf compile -q
```

The content of the `target` folder:

```bash
ls target
```

The "compiled" sql script which uses table defined in case `target.name == 'dev'` (default case).

```bash
cat target/compiled/models/script.sql
```

With `-t` parameter you can specify the target to use:

```bash
dbtf compile -q -t prod
```

Now script uses the value defined for `target.name == 'prod'`:

```bash
cat target/compiled/models/script.sql
```

## Run

To run dbt models use the commands `dbt run`, `dbtf run`. DBT automatically wraps your models in the required DDL (data definition language) and creates the corresponding table or view for each one.

---

Consider the example of running a dbt project. Following cell creates a project and defies the simple model in it.

```bash
#init
dbtf init --project-name knowledge --profile knowledge -q
cd knowledge
```

```bash
# file knowledge/models/script.sql
SELECT val1, val2
FROM (
    VALUES 
    (10, 20),
    (30, 40)
) AS temp(val1, val2)
```

Running:

```bash
dbtf run --select script -q
```

**Note**: `--select` is used to specify which model have to be used.

The following cell shows the database we are using as an example now have corresponding data.

```bash
PGPASSWORD=postgres psql -h localhost -U postgres -c "SELECT * FROM public.script"
```

## Show

The `dbt show` command allows you to preview the output of a given dbt model.

Check description in the [About dbt show commannd](https://docs.getdbt.com/reference/commands/show?version=2.0).

---

The following cells create a sample project and some model in it.

```bash
#init
dbtf init --project-name knowledge --profile knowledge -q
cd knowledge
```

```bash
# file knowledge/models/script.sql
SELECT 'some data' AS col
```

The output of the `dbt show` for the model:

```bash
dbt show -q --select script.sql
```

Fusion supports the `show` command as well:

```bash
dbtf show -q --select script.sql
```

### Inline

The `--inline` flag is sometimes relly usefull as it allows you to run a specified query in the target database.

---

The following cell initialises the DBT project and displays the databases of the PostgreSQL instanace that profiles points to with inline show.

```bash
#init
dbt init knowledge --profile knowledge -q
cd knowledge
dbt show --inline "select * from pg_database" -q
```

## List

The `dbt ls` (`dbt list`) command enables you to check the resources defined in the project.

The following features are available in the `ls` command:

- `--select` option to specify the model/group of models you want to check.
- Specifying `+` before or after allow you to view models that depend on your model and that your model depends on.

---

The following cell creates a dbt project and few models in it.

```bash
#init
dbt init knowledge --profile knowledge -q
cd knowledge
```

```bash
# file knowledge/models/model_a.sql
select 1;
```

Model `base_b` depends on the `base_a`:

```bash
#file knowledge/models/model_b.sql
select {{ ref('model_a') }};
```

The `model_c` depends on all previous models:

```bash
# file knowledge/models/model_c.sql
select * from {{ ref('model_b') }}
```

The output of the raw `dbt ls` command:

```bash
dbt ls -q
```

The models that somehow related to the `model_c`:

```bash
dbt ls -q --select +model_b+
```

All models that `model_c` depends on:

```bash
dbt ls -q --select +model_c
```
