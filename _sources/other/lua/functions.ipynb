# Functions

The functions in Lua are very similar to those in the other programming languages. However, this page considers some features that could be a bit surprising.

## Variable arguments

To define a function that takes an optional number of elements specify the `...` as the last argument. In the body of function refer to it using `...`. The values could be unpacked or turned into the table.

---

The following cell shows examples of how variable arguments can be defined and used inside a function:

```lua
function example_function(...)
    print(...)
    print({...})
    a, b = ...
    print(a, b)
end

example_function(10, 20, 'string literal')
```

Here is an example of variable arguments being mixed with regular arguments:

```lua
function example_function(a, b, ...)
    print(a)
    print(b)
    print(...)
end

example_function(10, 20, 'some extra', 'second extra')
```

## Default parameters

Lua **does not have** any special tools for defining the some arguments with default values.


However, there is a common workaround: check the function table to see if there are key-value pair exists and, it not, specify the default value.

---

The following cell defines the `default_param_examp` function, which demonstrates how the default values are processed within the function:

```lua
function default_param_examp(opts)
    if opts.default == nil then
        opts.default = 10
    end
    return opts.obligatory + opts.default
end

default_param_examp { obligatory = 7 }
```
