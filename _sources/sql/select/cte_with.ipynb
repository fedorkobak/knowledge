# CTE (WITH)

*Common Table Expressions* (CTEs) allow complex SQL queries to be broken down into separate temp queries within the code. The DB engine will substitutes them for the corresponding place during execution.

## Multiple

You can define several CTE's in one query. The syntax have to be like this:

```sql
WITH
name_one AS (<query>)
name_two As (<query>)

SELECT ...
```

---

The following cell defines two CTEs and joins them in the final query.

````sql
--postgreSQL
WITH
population AS (
    SELECT *
    FROM (
        VALUES
        ('Belarus', 10),
        ('Poland', 44),
        ('Ukraine', 54)
    ) AS temp(country, population)
),
capitals AS (
    SELECT *
    FROM (
        VALUES
        ('Belarus', 'Minsk'),
        ('Poland', 'Warsaw'),
        ('Ukraine', 'Kiev')
    ) AS temp(country, capital)
)
SELECT
    c.country, p.population, c.capital
FROM population AS p
    JOIN capitals AS c
    ON p.country = c.country;
````

## Recursive

A recursive CTE references to itself to return subset of data until all results are retrieved.

---

The following cell shows a simple recursive CTE that increments the result of the previous iteration.

````sql
--postgreSQL
WITH RECURSIVE numbers(n) AS (
    SELECT 1

    UNION ALL

    SELECT n + 1
    FROM numbers
    WHERE n < 5
)
SELECT *
FROM numbers;
````

When `n` reaches 5, the recursive member returns an empty output, so the execution stops.
