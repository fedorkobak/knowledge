# Types
At this page we'll consider different join types and their features in clickhouse and postgres.

## Full (outer)

The output of a full join satisfies the following statements:

* All rows from both tables are included.
* Rows that satisfy the matching condition are combined into a single row.
* Rows from either table that have no matching row in the other table are included, with `NULL` values in place of the other table’s columns.

---

The following cell demonstrates the application of the `FULL JOIN` in postgres. 

```sql
--postgreSQL
WITH 
table1(col1) AS (VALUES (1), (2), (3), (4)),
table2(col1) AS (VALUES (1), (3), (8), (9))

SELECT *
FROM 
    table1
    FULL JOIN table2 ON table1.col1=table2.col1
;
```

The following cell demonstrates the same example in ClickHouse; the result is nearly identical, with the only difference being that it uses zeros instead of NULLs for non-matching rows.

```sql
--ClickHouse
CREATE TEMPORARY TABLE table1(val Int32) ENGINE = Memory;
CREATE TEMPORARY TABLE table2(val Int32) ENGINE = Memory;

INSERT INTO table1 VALUES
    (1), (2), (3), (4);
INSERT INTO table2 VALUES
    (1), (3), (8), (9);

SELECT table1.val, table2.val
    FROM table1 
    FULL JOIN table2 ON table1.val=table2.val;
```

## Cross

A cross join results in the Cartesian product of the input tables (rus. декартово произведение): each combination of rows from the input tables rows is included in the output. This type of join doesn't require the condition a to be specified in `ON` clause.

---

The following cell displays the results of the cross-joining two arrays of numbers.

```sql
--postgreSQL
WITH 
table1(col1) AS (VALUES (1), (2), (3), (4)),
table2(col1) AS (VALUES (1), (3), (8), (9))

SELECT *
FROM 
    table1
    CROSS JOIN table2
;
```

## Self

A self join is a type of join that joins a table with itself.

---

The cell defines the example table and joins each row with another row whose value is exactly one less.

```sql
--postgreSQL
WITH 
example(col1) AS (VALUES (1), (2), (3), (4))

SELECT * FROM example t1 JOIN example t2 ON t1.col1 = t2.col1 + 1;
```
