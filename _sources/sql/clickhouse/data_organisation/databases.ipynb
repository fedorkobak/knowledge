# Databases

The database concept in ClickHouse allows you to logically separate tables belonging within one ClickHouse instance. This page covers core concepts involed in manipulating the ClickHouse databases. 

## Tables

To display tables from a specific database, use the command `SHOW TALES FROM <database name>;` or `SHOW TABLES IN <database name>`.

---

The two options to display the list of tables in the database are shown in the following cells.

```sql
--ClickHouse
SHOW TABLES FROM information_schema LIMIT 5;
```

```sql
--ClickHouse
SHOW TABLES IN information_schema LIMIT 5;
```

### Current database

In the ClickHouse session, there is a current database. The tables in this database will be used by default in the quieries by defaul't. 

- **Show** current database with: `currentDatabase()`.
- To change current database: `USE <database name>;`.

---

The following cell shows the current database - `deafult`.

```sql
--ClickHouse
SELECT currentDatabase();
```
