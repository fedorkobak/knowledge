# Tesseract

Is a popular opensource tool for OCR this page considers some features of the technology.

## Layout

Tesseract allows to fetch information about the layout of the document. In CLI interface use options:

- `tsv` to get data in tab-separated format.
- `hocr` to get data in special html-like hocr format.

---

Consider the following image as the input example:

![](tesseract_files/layout_example.png)

The regular tesseract call:

```bash
tesseract tesseract_files/layout_example.png stdout
```

The `tsv` call:

```bash
tesseract tesseract_files/layout_example.png stdout tsv | head -n 10
```

Each piece of text has additional information describing its possition.

And the same example for `hocr` output:

```bash
tesseract tesseract_files/layout_example.png stdout hocr | head -n 20
```

These documents can be viewed using standard web browser.
