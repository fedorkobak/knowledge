# OpenSSL

This page considers the `openssl` CLI.

## Asymmetric encryption

The set of commands usefull for asymmetric encryption:

- `openssl genkey`: for generating the keys.
- `openssl pkey`: to extrac the public key from the private.
- `openssl pkeyutl -encrypt`: to encrypt the data.
- `openssl pkeyutl -decrypt`: to decrypt the data.

---

Consider all the operations mentioned earlier, providing examples for each one.

Generation of the private key and extraction of the public key:

```bash
# init
openssl genpkey -algorithm RSA -out private.pem
```

```bash
openssl pkey -in private.pem -pubout -out public.pem
ls
```

For example, let's take a file containing data for encryption. **Note**: to use the public key for encryption, set up the `-pubin` parameter.

```bash
echo "data for encryption" > encrypt.data
openssl pkeyutl -encrypt -pubin -inkey public.pem -in encrypt.data -out encrypted.data
```

The following cell shows the encrypted file.

```bash
cat encrypted.data
```

Decryption of the file and its representation:

```bash
openssl pkeyutl -decrypt -inkey private.pem -in encrypted.data -out decrypted.data
cat decrypted.data
```

## Authenticity

Authenticity is the particular case of the asymmetric enryption, where the private key is used for encryption and the public key for decryption. This proofes that the information was encoded by the owner of the private key. The different parameters of the `openssl pkeyutl` are used for this operation:

- `openssl pkeyutl -sign` to encrypt the information.
- `openssl pkeyutl -verifyrecover` to decrypt the information.
