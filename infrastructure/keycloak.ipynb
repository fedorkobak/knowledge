## Service account

Service account is used to configure the authentification of other services in the resources protected by Keycloak.

Check more in [Using a service account](https://www.keycloak.org/docs/latest/server_admin/index.html#_service_accounts) sectoin of the documentation.

### Admin

The admin account by default is bound to the `admin-cli` client by default. This client has "Direct access grants", it is a service account.

---

The following cell shows how to obtain an access token for an admin account.

```bash
curl -X POST -s \
    -d "grant_type=password"\
    -d "client_id=admin-cli" \
    -d "username=admin" \
    -d "password=admin" \
    http://localhost:8080/realms/master/protocol/openid-connect/token | jq "."
```

There is a shortcut in utils for this operation:

```bash
get_admin_token | jq
```

## Users

Keycloak handles users. The endpoint that is responsible for retrieving user information is [/admin/realms/{realm}/users](https://www.keycloak.org/docs-api/latest/rest-api/index.html#_get_adminrealmsrealmusers).

---

The following cell gets the id of the `admin` user.

```bash
access_token=$(get_admin_token | jq -r .access_token)

curl -X GET -s \
    -H "authorization: Bearer $access_token" \
    http://localhost:8080/admin/realms/master/users?username=admin | jq .[0].id
```

This notebook uses `get_admin_user` shortcut to access the information about admin user:

```bash
get_admin_user | jq | head -n 20
```

## Sessions

A session is an entity that stores user login information: the user's identity, the time the session started and the last access time.

- A session is **created** each time you request an access token. The `session_state` attribute of the token request contains the id of the corresponding session.

- You can the information about sessions using following:

    - [GET `/admin/realms/{realm}/users/{user-id}/sessions`](https://www.keycloak.org/docs-api/latest/rest-api/index.html#_get_adminrealmsrealmusersuser_idsessions) sessions of the particular user.
    - [GET `/admin/realms/{realm}/clients/{client-uuid}/user-sessions`](https://www.keycloak.org/docs-api/latest/rest-api/index.html#_get_adminrealmsrealmclientsclient_uuiduser_sessions) sessions of the given client.

- You can delete the given session using [DELETE `/admin/realms/{realm}/sessions/{sessions}`](https://www.keycloak.org/docs-api/latest/rest-api/index.html#_delete_adminrealmsrealmsessionssession).

---

The following cell requests an access token, which creates a new session. The next cell shows the corresponding session ID.

```bash
token_data=$(get_admin_token)
access_token=$(echo $token_data | jq -r .access_token)
session_state=$(echo $token_data | jq -r .session_state)

echo $session_state
```

The next cell shows a list of sessions for the admin user.

```bash
admin_id=$(get_admin_user | jq -r .id)

curl -X GET -s \
    -H "authorization: Bearer $access_token" \
    http://localhost:8080/admin/realms/master/users/${admin_id}/sessions | jq '.[0:2]'
```

**Note** it should be at least one because at the begining script requests an admin token at the begining.

The following cell deletes the session that corresponds to the token we used for authorization:

```bash
curl -X DELETE  \
     -H "Authorization: Bearer $access_token" \
     "http://localhost:8080/admin/realms/master/sessions/$session_state"
```

As a result, any request using this token as a credential will fail due to authorization error.

```bash
curl -X GET -s -H "authorization: Bearer $access_token" \
    http://localhost:8080/admin/realms/master/users/count
```
