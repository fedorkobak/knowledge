# Keycloak

### Service account

Service account is used to configure the authentification of other services in the resources protected by Keycloak.

Check more in [Using a service account](https://www.keycloak.org/docs/latest/server_admin/index.html#_service_accounts) sectoin of the documentation.

#### Admin

The admin account by default is bound to the `admin-cli` client by default. This client has "Direct access grants", it is a service account.

---

The following cell shows how to obtain an access token for an admin account.

```bash
curl -X POST -s \
    -d "grant_type=password"\
    -d "client_id=admin-cli" \
    -d "username=admin" \
    -d "password=admin" \
    http://localhost:8080/realms/master/protocol/openid-connect/token | jq "."
```

### Users

Keycloak handles users. The endpoint that is responsible for retrieving user information is [/admin/realms/{realm}/users](https://www.keycloak.org/docs-api/latest/rest-api/index.html#_get_adminrealmsrealmusers).

---

The following cell gets the id of the `admin` user.

```bash
access_token=$(curl -X POST -s \
    -d "grant_type=password"\
    -d "client_id=admin-cli" \
    -d "username=admin" \
    -d "password=admin" \
    http://localhost:8080/realms/master/protocol/openid-connect/token | jq -r .access_token)

curl -X GET -s \
    -H "authorization: Bearer $access_token" \
    http://localhost:8080/admin/realms/master/users?username=admin | jq .[0].id
```
