# Random distributions

In `torch` there is a set of functions that allow you to generate tensors with random values distributed according to the given distribution.

```python
import torch

import seaborn as sns
import matplotlib.pyplot as plt
```

## rand

The result will be randomly generated values with a uniform distribution from 0 to 1. The following graph shows the histogram for generation.

```python
sns.histplot(torch.rand(10000))
plt.show()
```

## randn

`torch.randn` generates random values normally distributed with a mean of 0 and a standard deviation of 1.

```python
sns.histplot(torch.randn(10000))
plt.show()
```
